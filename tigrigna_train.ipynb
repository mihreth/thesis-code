```python
from transformers import RobertaConfig

config = RobertaConfig(
    vocab_size=52_000,
    max_position_embeddings=514,
    num_attention_heads=12,
    num_hidden_layers=6,
    type_vocab_size=1,
)
```

```python
tokenizer_folder = '../../tokenizer/tokenized/tigrigna' # tokenizer folder
```

```python
from transformers import RobertaTokenizerFast

tokenizer = RobertaTokenizerFast.from_pretrained(tokenizer_folder, max_len=512)
```

```python
tokenizer('áˆ°áˆ‹áˆ áŠ¨áˆ˜á‹­ áŠ£áˆˆáŠ»á¢')
```

```python
from transformers import RobertaForMaskedLM

model = RobertaForMaskedLM(config=config)
```

```python
model.num_parameters()
# => 84 million parameters
```

```python
from transformers import LineByLineTextDataset

dataset = LineByLineTextDataset(
    tokenizer=tokenizer,
    file_path="../../datasets/unlabeled/tigrigna.txt",
    block_size=128,
)
```

```python
from transformers import DataCollatorForLanguageModeling

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer, mlm=True, mlm_probability=0.15
)
```

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir="../../trained_model/tigrigna",
    overwrite_output_dir=True,
    num_train_epochs=8,
    per_device_train_batch_size=8,
    save_steps=10_000,
    save_total_limit=2,
    prediction_loss_only=True,
    resume_from_checkpoint=True
)

trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=dataset,
)
```

```python
trainer.train()
```
